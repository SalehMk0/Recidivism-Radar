```python
cases = pd.DataFrame({
    'risk_decile_score': [1, 5, 3, 4, 8, 6, 10, 9, 2, -1, 7],
    'risk_score_category': ['Low', 'Medium', 'Low', 'Low', 'High', 'Medium', 'High', 'High', 'Low', 'No Data', 'Medium']
})


def map_risk_score(score):
    if score == -1:
        return 'No Data'
    elif 1 <= score <= 4:
        return 'Low'
    elif 5 <= score <= 7:
        return 'Medium'
    elif 8 <= score <= 10:
        return 'High'
    else:
        return 'No Data'


cases['risk_score_category'] = cases['risk_decile_score'].apply(map_risk_score)


d_risk_score_df = pd.DataFrame({
    'risk_decile_score': cases['risk_decile_score'],
    'risk_score_category': cases['risk_score_category']
})


d_risk_score_df['id'] = range(1, len(d_risk_score_df) + 1)


d_risk_score_df.to_csv('d_risk_score.csv', index=False)
```
