```python

charge_categories = {
    'Violent Crimes': ['Aggravated Assault', 'Battery', 'Robbery', 'Kidnapping', 'Murder', 'Manslaughter'],
    'Drug Offenses': ['Possession', 'Trafficking', 'Deliver', 'Purchase', 'Manufacture'],
    'Property Crimes': ['Burglary', 'Theft', 'Trespass', 'Criminal Mischief', 'Arson', 'Possession of burglary tools'],
    'Traffic Offenses': ['DUI', 'Driving', 'License', 'Fleeing', 'Eluding'],
    'Public Order Offenses': ['Disorderly', 'Loitering', 'Prostitution'],
    'White Collar Crimes': ['Fraud', 'Forgery', 'Counterfeiting', 'Embezzlement'],
    'Sexual Offenses': ['Sexual', 'Lewd', 'Lascivious'],
    'Weapons Offenses': ['Firearm', 'Weapon', 'Concealed'],
    'Other': []  
}


def categorize_charge(charge):
    if pd.isna(charge): 
        return 'Other'
    charge = str(charge).lower()  
    for category, keywords in charge_categories.items():
        if any(keyword.lower() in charge for keyword in keywords):
            return category
    return 'Other'

df['charge_category'] = df['current_charge_description'].apply(categorize_charge)
```

```python
def categorize(row):
    if row['is_recidivist'] == 0 and row['is_violent_recidivist'] == 0 and row['risk_decile_score'] < 5 and row['recidivism_risk_decile_score'] < 5:
        return 'Likely a Rehabilitation'
    elif row['is_recidivist'] == 1 or row['is_violent_recidivist'] == 1 or row['risk_decile_score'] >= 5 or row['recidivism_risk_decile_score'] >= 5:
        return ' Likely a Punishment'
    else:
        return 'Uncategorized'
```
